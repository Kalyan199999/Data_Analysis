# Dataset Link
# https://www.kaggle.com/datasets/rishikeshkonapure/zomato/data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
path = '../datasets/zomato.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.drop(columns=['url','address'] , axis=1,inplace=True)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

# Rename the columns

```python
new_names = {
    'approx_cost(for two people)':'cost_2_people',
    'listed_in(type)':'type',
    'listed_in(city)':'city'
}
df.rename(new_names , axis=1 ,inplace=True)
```

```python
df.head(3)
```

# Data Cleaning

```python
df['rate'] = df['rate'].replace('NEW', np.nan)
df['rate'] = df['rate'].replace('-', np.nan)
# df['rate'] = df['rate'].replace('nan', np.nan)
df['rate'] = df['rate'].str.replace('/5', '').astype(float)
```

```python
df['cost_2_people'] = df['cost_2_people'].str.replace( ',' , '' ).astype(float)
```

```python
df.shape
```

# remove the duplicates and the null values

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df.shape
```

# Online and table booking orders

```python
online_offline = list(df['online_order'].value_counts())
tales_book = list( df['book_table'].value_counts() )


fig, axes = plt.subplots(1, 2, figsize=(10, 3))

axes[0].pie(x=online_offline,labels=['Online Orders','Offline Orders'],autopct='%1.1f%%' , colors=['#99ff99', '#ffcc99'])
axes[0].set_title('Online order availability')
axes[0].axis('equal')

axes[1].pie(x=tales_book, labels=['No','Yes'],autopct='%1.1f%%', colors=['#66b3ff', '#fff999'])
axes[1].set_title('Table Booking Availability')
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

# Resturent Types

```python
df['rest_type'] = df['rest_type'].apply(
    lambda x: ', '.join(sorted(x.split(', '))) if isinstance(x, str) else x
)
```

```python
rest_types = dict( df['rest_type'].value_counts().head(10) )

x = list(rest_types.values())
y = list(rest_types.keys())

plt.figure(figsize=(10, 5))
plt.bar(x=y, height=x, color='skyblue')
plt.xlabel("Restaurant Type")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.title("Distribution of Restaurant Types")
plt.tight_layout()
plt.show()
```

# Ratings Curve

```python
ratings = df.groupby('rate')
rating_dict = dict( ratings['name'].count() )
```

```python
# Convert keys and values to lists
labels = list(rating_dict.keys())
counts = list(rating_dict.values())

# Create numeric positions for each bar
x = np.arange(len(labels))

# Plot with bar width < 1 to create gaps
plt.figure(figsize=(10, 4))
plt.bar(x, counts, width=0.6, color='skyblue', edgecolor='black')

# Set the x-ticks and labels
plt.xticks(x, labels, rotation=45)

# Add grid and labels
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.title("Ratings Plot")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

# frequency of Resturents at the location

# Top 20 Resturents names with votes

```python
rest_names = df.groupby('name')
```

```python
resurent_votes = rest_names['votes'].sum()
```

```python
sorted_votes = resurent_votes.sort_values(ascending=False)
sorted_votes = sorted_votes.head(20)
sorted_votes = dict(sorted_votes)
```

```python
plt.figure(figsize=(20,5))
plt.title("Top 20 Resturents names with votes")
plt.bar( x=sorted_votes.keys() , height=sorted_votes.values() ,color='lightcoral' )
plt.xlabel("Resturent name")
plt.ylabel("Frequency")
plt.xticks( rotation=75)
plt.show()
```

# Top 20 resturnts with avg cost for 2 people

```python
resurent_avg_cost = rest_names['cost_2_people'].mean()
sorted_avg_cost = resurent_avg_cost.sort_values(ascending=False)
sorted_avg_cost = sorted_avg_cost.head(20)
sorted_avg_cost = dict(sorted_avg_cost)
```

```python
plt.figure(figsize=(20,5))
plt.title("Top 20 Resturents names with its cost for 2 people")
plt.bar( x=sorted_avg_cost.keys() , height=sorted_avg_cost.values() ,color='orange' )
plt.xlabel("Resturent name")
plt.ylabel("Cost")
plt.xticks( rotation=75)
plt.show()
```
